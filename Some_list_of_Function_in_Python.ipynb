## List down of function list of if , else, and elif condition

Here are some commonly used functions in Python for if, else and elif conditionals:

1. `if` statement: The `if` statement is used to check whether a given condition is true or false. If the condition is true, the code block inside the `if` statement will be executed.

Example:
```
x = 5
if x < 10:
    print("x is less than 10")
```

2. `else` statement: The `else` statement is used in conjunction with the `if` statement to execute a block of code when the condition in the `if` statement is false.

Example:
```
x = 15
if x < 10:
    print("x is less than 10")
else:
    print("x is greater than or equal to 10")
```

3. `elif` statement: The `elif` statement is used to add an additional condition to the `if` statement. It is used in between the `if` and `else` statements.

Example:
```
x = 15
if x < 10:
    print("x is less than 10")
elif x > 20:
    print("x is greater than 20")
else:
    print("x is between 10 and 20")
```

4. `and` operator: The `and` operator is used to combine multiple conditions in an `if`, `elif` or `while` statement. It returns a boolean value of `True` if both conditions are true, and `False` otherwise.

Example:
```
x = 5
y = 10
if x < 10 and y > 5:
    print("Both conditions are true")
```

5. `or` operator: The `or` operator is used to combine multiple conditions in an `if`, `elif` or `while` statement. It returns a boolean value of `True` if at least one of the conditions is true, and `False` if both conditions are false.

Example:
```
x = 5
y = 10
if x < 10 or y < 5:
    print("At least one of the conditions is true")
```

Note that the above examples are just a few of the many functions available in Python for conditionals.

##List down a full list of functions of for loop  for python programming

A "for loop" in Python is a type of loop that iterates over a sequence of values. Here's a list of functions that you can use with for loops in Python:

1. `range()`: generates a sequence of integers starting from 0 by default. It takes up to three arguments: start, stop, and step.

2. `enumerate()`: returns a tuple containing a count for each iteration and the values obtained from the iterable. It adds a counter to an iterable and returns it in a form of an enumerate object.

3. `zip()`: Takes two or more iterables as input and zips them together, creating a new iterable with pairs of corresponding elements from those input iterables.

4. `len()`: returns the number of items in an iterable.

5. `sorted()`: returns a new sorted list from the items in an iterable.

6. `reversed()`: returns an iterator that yields the items from an iterable in reverse order.

7. `break`: terminates the loop immediately.

8. `continue`: skips the current iteration of the loop and continues to the next one.

9. `else`: specifies a block of code to be executed when the loop finishes without hitting any breaks.

10. `pass`: a null operation — when it is executed, nothing happens.

These are some of the commonly used functions and control statements that you can use with for loops in Python. You can find more in the official Python documentation.
